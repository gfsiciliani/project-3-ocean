## Import Libraries

```python
# Import the required libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pathlib import Path
import seaborn as sns
sns.set_theme(style="darkgrid")

# Create a reference to the CSV and import it into a Pandas DataFrame
cubs = Path("seal_cub_count_data/Antarctic_Fur_Seal_Pup_Counts_CS_1995_2012_v1_0.csv")
cubs_df = pd.read_csv(cubs)
cubs_df.head()
```

```python
cubs_df[['Start of Season', 'End of Season']] = cubs_df['Season'].str.split('/', expand=True)
print(cubs_df)
```

```python
sns.lmplot(data = cubs_df, x='End of Season', y='Pup Count')
plt.show()
```
