```python
import pandas as pd
```

```python
df = pd.read_csv ('storms.csv')
df.head()
```

```python
catfive_df = df[df['category'] == 5]

# Display the resulting data frame
catfive_df.head()
```

```python
df_no_duplicates = catfive_df.drop_duplicates(subset='name', keep='first')

# Display the resulting data frame
df_no_duplicates.head(100)
```

```python
#csv came from https://www.kaggle.com/datasets/utkarshx27/noaa-atlantic-hurricane-database
```

```python
columns_to_drop = ['tropicalstorm_force_diameter', 'hurricane_force_diameter']

# Drop the specified columns
final_df = df_no_duplicates.drop(columns=columns_to_drop)
final_df
```

```python
#Create data frame to include all data points for category 5 hurricanes
print(final_df['name'])
target_names = final_df['name'].tolist()
```

```python
# target_names = ['Anita','David','Allen','Gilbert','Hugo','Andrew','Mitch','Isabel','Ivan','Emily','Katrina','Rita','Wilma','Dean','Felix','Matthew','Irma','Maria','Michael','Dorian','Lorenzo']
# target_names = ['Anita','David','Allen','Gilbert','Hugo','Andrew','Mitch','Isabel',               'Katrina','Rita','Wilma',               'Matthew','Irma',                  'Dorian'          ]
filtered_df = df[df['name'].isin(target_names)]

# Display the filtered DataFrame
filtered_df.head(25)
```

```python
columns_to_drop = ['tropicalstorm_force_diameter', 'hurricane_force_diameter','category']

# Drop the specified columns
final_df = filtered_df.drop(columns=columns_to_drop)
final_df.head()
```

```python
target_status = ['hurricane']
hurricane_status_df = final_df[df['status'].isin(target_status)]
hurricane_status_df.head(20)
len(hurricane_status_df)
```

```python
hurricane_status_df.to_csv('hurricanes.csv', index=False)
```
